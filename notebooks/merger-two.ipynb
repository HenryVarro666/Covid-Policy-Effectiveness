# Creating train2.csv : regression models 

`Only working with state wise policies`



```python
import pandas as pd
```

```python
policyDF = pd.read_csv('../outputs/allPolicy.csv',parse_dates=['date_implement'])

trainDF  = pd.read_csv('../outputs/train.csv')
trainDF = trainDF.dropna()
```

# Creating data for big 

```python
policyDF_national = policyDF[policyDF['state']=='national']
policyDF_statewise = policyDF[policyDF['state']!='national']
```

```python
# Adding Policy type as 0 - state policy 
policyDF_statewise['policy_type']=0

# Adding Policy type as 1 - national policy 
policyDF_national['policy_type'] = 1
```

```python
# Merging state policy change points with their profiles
policyDF_statewise_FULL = policyDF_statewise.merge(trainDF,left_on='state',right_on='stateName',how='left')
```

```python
policyDF_statewise_FULL.to_csv('../outputs/train2_state.csv')
```
