# Some fun analytics 

```python
import pandas as pd
import seaborn as sb
pd.set_option('display.max_columns', None)
from scipy.stats import zscore
from sklearn.cluster import KMeans

```

```python
trainDF = pd.read_csv('../outputs/train.csv')
trainDF = trainDF.dropna()
```

```python
# Finalizing Indexes so we can drop state names 
trainDF = trainDF.reset_index(drop=True)
featureColumns = trainDF.columns.to_list()
```

```python
featureColumns.remove('stateName')
featureColumns.remove('CEN_stateCode')
featureColumns.remove('Unnamed: 0')
featureColumns.remove('CEN_Unnamed: 0')
featureColumns.remove('CEN_Unnamed: 0.1')
```

```python
stateDF = trainDF[['stateName','CEN_stateCode']]
featureDF = trainDF[featureColumns]
```

```python

featureDF_zscore = featureDF.apply(zscore)
featureDF_zscore.head()
```

# Correlation between the columns

Basically what i got is `obviousness is really obvious`

```python
def easeOfVisHeavy(corScore):
    if corScore>0.7:
        return 1 
    elif corScore<-0.7:
        return -1
    else:
        return 0 
    
def easeOfVisMid(corScore):
    if corScore>0.5:
        return 1 
    elif corScore<-0.5:
        return -1
    else:
        return 0 
    
def easeOfVisLow(corScore):
    if corScore>0.3:
        return 1 
    elif corScore<-0.3:
        return -1
    else:
        return 0 
    
```

## Heavy Correlation

```python
featureDF_zscore.corr().applymap(easeOfVisHeavy)
```

## Mid Correlation

```python
featureDF_zscore.corr().applymap(easeOfVisMid)
```

## Low Correlation

```python
featureDF_zscore.corr().applymap(easeOfVisLow)
```
