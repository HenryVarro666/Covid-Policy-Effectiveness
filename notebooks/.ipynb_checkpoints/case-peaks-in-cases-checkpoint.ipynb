# Finding Peak times in each state
We hypothized that the month, each state recieved a peak is a good profile for that state ie. 
> States which have the peak appear at similar times share somewhat of a similar charactaristic

Insights:
- We do not use original case data as it is noisy and hard to find slopes in. 
- We use interpolation of data which we previously created. 
- Manual Clearning (Show outputs)


This notebook works with the main goal of finding which months did the peak occur in each state. 

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = [18, 10]
import numpy as np
import seaborn as sns
from random import randrange
import matplotlib.pyplot as plt
from scipy.misc import electrocardiogram
from scipy.signal import find_peaks


```

```python
case_df = pd.read_csv('../outputs/case_processed.csv')
states = case_df['state'].unique()
```

## 1. Experimenting on One State

####  Run from here again and again to get different states at the end 

```python
# Pick a state randomly 
random_state_picker = randrange(len(states))
```

```python
target_df = case_df[case_df['state']==states[random_state_picker]]

# Resetting index for easy manipulation 
target_df = target_df.reset_index(drop=True)

print(f'State picked :{states[random_state_picker]} ')
```

```python
target_df.plot(y=['caseInterpolate_gauss8','new_case'],x='submission_date')
```

```python
# The curve we find the peaks in 
target_df.plot(y=['caseInterpolate_gauss8'],x='submission_date')
```

```python

peaks, properties = find_peaks(target_df['caseInterpolate_gauss8'].to_numpy(), height=0)

sorted_peaks_by_height = [x for _, x in sorted(zip(properties['peak_heights'], peaks))][::-1]

for peak in sorted_peaks_by_height[:4]: 
    target_df.loc[peak.item(), 'peakValue'] = target_df.loc[[peak]]['caseInterpolate_gauss8'].to_list()[0]
```

```python
ax = target_df.plot(y=['peakValue'],x='submission_date', color='red', marker='o',
     markerfacecolor='red', markersize=12)
target_df.plot(y=['caseInterpolate_gauss8'],x='submission_date', ax=ax)

plt.show()
plt.savefig('../outputs/peak_visualisations/')
```

```python
# Dates that 
for peakdates in sorted_peaks_by_height:
    print(target_df.loc[[peakdates]]['submission_date'].to_list()[0])
```

```python
FINAL_OP = []
for state in states:
    
    # Selecting the state 
    target_df = case_df[case_df['state']==state]
    print(f'State Selected : {state}')
    # Reset index for better numbering
    target_df = target_df.reset_index(drop=True)
    
    # Finding Peaks
    peaks, properties = find_peaks(target_df['caseInterpolate_gauss8'].to_numpy(), height=0)

    # Sorting Peaks and Properties together wrt height
    sorted_peaks_by_height = [x for _, x in sorted(zip(properties['peak_heights'], peaks))][::-1]
    print(f'peaks found index : {sorted_peaks_by_height}')
    sorted_peaks_by_heigh_properties = properties['peak_heights'].tolist()
    
    if len(sorted_peaks_by_height) == 0 :
        print('No Peaks Found')
        FINAL_OP.append({
            'state': state,
            'peak_dates': [],
            'peak_properties' : []
        })
        continue
    else: 
        for peak in sorted_peaks_by_height: 
            # Adding the peakvalue variable
            target_df.loc[peak, 'peakValue'] = target_df.loc[[peak]]['caseInterpolate_gauss8'].to_list()[0]
            
            
        ax = target_df.plot(y=['peakValue'],x='submission_date', color='red', marker='o',markerfacecolor='red', markersize=12)
        target_df.plot(y=['caseInterpolate_gauss8'],x='submission_date', ax=ax)
        plt.savefig(f'../outputs/peak_visualisations/{state}.pdf')

        peak_date_array = []
        for peakdates in sorted_peaks_by_height:
            peak_date_array.append(target_df.loc[[peakdates]]['submission_date'].to_list()[0])
            pass

        FINAL_OP.append({
            'state': state,
            'peak_dates': peak_date_array,
            'peak_properties' : sorted_peaks_by_heigh_properties
        })
        
OUTPUT_DF = pd.DataFrame(FINAL_OP)
OUTPUT_DF.to_csv('../outputs/state_peak.csv')

    
```
